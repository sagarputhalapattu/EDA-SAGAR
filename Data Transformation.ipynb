```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
visa_df = pd.read_csv('Visadataset.csv')
visa_df.head(2)
```

```python
exp_data = np.random.exponential(size=10000)
exp_data
```

```python
plt.hist(exp_data,bins=20,label='EXP')
plt.show()
```

```python
nor_data = np.random.normal(size=1000)
nor_data
```

```python
plt.hist(nor_data,bins=20,label='NORMAL')
plt.show()
```

**main Goal**

- convert Exp_data to Normal Distribution

$Step-4$

- log transformation mean performing logarithm opretion on origianal data

-  it  is an approch to convert into normal distribution

-   log mean natural logarithm base = e

```python
log_data = np.log(exp_data)
plt.hist(log_data,bins=20,label='NORMAL')
plt.show()
```

```python
plt.figure(figsize=(10,4))
plt.subplot(1,2,1).hist(exp_data,bins=30,label="EXP")
plt.legend()
plt.subplot(1,2,2).hist(log_data,bins=30,label="LOG")
plt.legend()
plt.show()
```

**Reciprocal transFromation**



```python
 x = 2 

1/x
```

```python
res_data = np.reciprocal(exp_data)
```

```python
plt.figure(figsize=(10,4))
plt.subplot(1,3,1).hist(exp_data,bins=30,label="EXP")
plt.legend()
plt.subplot(1,3,2).hist(log_data,bins=30,label="LOG")
plt.legend()
plt.subplot(1,3,3).hist(res_data,bins=30,label="RES")
plt.legend()
plt.show()
```

**squrt>root**

```python
sq_root = np.sqrt(exp_data)
```

```python
plt.figure(figsize=(10,4))
plt.subplot(1,3,3).hist(sq_root,bins=30,label='SQr')
plt.legend()
plt.subplot(1,3,2).hist(log_data,bins=30,label="LOG")
plt.legend()
plt.subplot(1,3,1).hist(exp_data,bins=30,label="EXP")
plt.legend()
plt.show()
```

**power-transformtion**

```python
visa_df = pd.read_csv('Visadataset.csv')
from sklearn.preprocessing import PowerTransformer
pt = PowerTransformer(method='yeo-johnson',)
data1 = exp_data.reshape(-1,1)
pt.fit_transform(data1)
plt.hist(data1)
plt.show()
```

```python
import numpy as np
from sklearn.preprocessing import PowerTransformer
pt = PowerTransformer(method='box-cox')
data = exp_data.reshape(-1,1)
pt.fit_transform(data)
plt.hist(data)
plt.show()
```
