```python
l1=[]
l2=[]
con= visa_df['continent'].unique()
for i in con:
    con_1 =(visa_df['continent']==i)
    con_2 = visa_df['case_status']=='Certified'
    con_3 = visa_df['case_status']=='Denied'
    certi_con = con_1&con_2
    denied_con = con_1&con_3
    l1.append(len(visa_df[certi_con]))
    l2.append (len(visa_df[denied_con]))

pd.DataFrame(zip(l1,l2),columns=['certified','Denied'],index=con)



 


    
```

**cross tab**

  -  cross tab 

```python
col1 = visa_df['continent']
col2 = visa_df['case_status']
result1=pd.crosstab(col1,col2)
result1
```

```python
result1.plot(kind = 'bar')
```

```python
col1 = visa_df['continent']
col2 = visa_df['case_status']
result2=pd.crosstab(col1,col2).T
result2.plot(kind = 'bar')
```

```python
col1 = visa_df['continent']
col2 = visa_df['education_of_employee']
col3 = visa_df['case_status']
col = [col1,col2]
result3=pd.crosstab(col,col3)
r3 = result3.plot(kind='bar')

r3


```
